```python
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('./images/nature1.jpg')
```

```python
imshow("Original",image)
```

### cv2.split function is used to split the image into corresponding 3 color segments with a B/W representation where the i

```python
B, G, R = cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```

```python
imshow("Blue Channel Only", B)
```

## Now in order to display the color scheme of each component separately we will need to zero out the other color components of the image.
 
###  Now we use the following function in order to create a matrix of zeros with the dimensions of the image<span style="color:blue"> The code creates a matrix with the first two dimensions of the image which contains its height and weight and not take the third component which contains the color segements </span>

```python
zeros = np.zeros(image.shape[:2],dtype="uint8")
```

```python
imshow("Red",cv2.merge([zeros,zeros,R]))
```

```python
imshow("Green",cv2.merge([zeros,G,zeros]))
```

```python
imshow("Blue",cv2.merge([B,zeros,zeros]))
```

## We can also use the above way in modifing the color segements separately in order to get a more intense red or a less green version of the image


```python
imshow("Red Boosted",cv2.merge([B,G,R+100]))
```

```python
imshow("Green deintensified",cv2.merge([B,G-200,R]))
```

## <span style="color:red">HSV COLOR SPACE</span>

```python
hsvImage = cv2.imread('./images/HSV_color_solid_cone.jpg')
imshow("HSV Color Space",hsvImage)
```

```python
image = cv2.imread('./images/nature1.jpg')

# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
imshow('HSV', hsv_image)
```

### Here we can see that after converting the image in HSV format it is difficult to interpret the image as openCV only works in reading the image in RGB format

```python
# Switching back to viewing the RGB representation
imshow("Hue", hsv_image[:, :, 0])
imshow("Saturation", hsv_image[:, :, 1])
imshow("Value", hsv_image[:, :, 2])
```
