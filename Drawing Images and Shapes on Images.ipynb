````python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
````

## Now we begin by creating a blank canvas, one with all three dimensions and one just being grayscale

````python
# Create a black image using numpy to create and array of black 
image = np.zeros((512,512,3), np.uint8)

# Can we make this in black and white? grayscale
image_gray = np.zeros((512,512), np.uint8)

# Black would be the same as a greyscale or color image (same for white)
imshow("Black Canvas - RGB Color", image,5)
imshow("Black Canvas - Grayscale", image_gray,5)
````

### **Let's draw a line over our black square**

```cv2.line(image, starting cordinates, ending cordinates, color, thickness)```

<span style="color:red">Note that unlike other cv2 functions, this is an inplace function meaning it makes changes to the original image</span>


````python
cv2.line(image, (0,0), (511,511), (255,127,0), 5)

imshow("Black Canvas With Diagonal Line", image,5)
````

````python
cv2.line(image, (0,511), (511,0), (125,127,0), 5)
imshow("Black Canvas With Diagonal Line", image,5)
````

## Now we move onto shapes like rectangles and triangles

<ol>
    <li><span style="color:red">cv2.rectangle</span> <span style="color:blue">(image, starting vertex, opposite vertex, color, thickness)</span></li>
    <li><span style="color:red">cv2.circle </span> <span style="color:blue"> (image, center, radius, color, fill)</span></li>
    <li><span style="color:red">cv2.polylines</span> <span style="color:blue">(image, points, Closed?, color, thickness)</span> if Closed = True, we join the first and last points.</li>
    </ol>

````python
# Create our black canvas again because now it has a line in it
image = np.zeros((512,512,3), np.uint8)

# Thickness - if positive. Negative thickness means that it is filled
cv2.rectangle(image, (100,100), (300,250), (127,50,127), 10)
imshow("Black Canvas With Pink Rectangle", image,4)
````

````python
image = np.zeros((512,512,3), np.uint8)

cv2.circle(image, (250, 350), 100, (15,150,50), 1) 
imshow("Black Canvas With Green Circle", image,4)
````

````python
image = np.zeros((512,512,3), np.uint8)

# Let's define four points
pts = np.array( [[104,50], [400,50], [90,200], [50,500]], np.int32)

# Let's now reshape our points in form  required by polylines
pts = pts.reshape((-1,1,2))

cv2.polylines(image, [pts], True, (0,0,255), 3)
imshow("Black Canvas with Red Polygon", image)
````

### **And now for adding text with cv2.putText**

```cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)```

**Available Fonts**
- FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX,FONT_HERSHEY_COMPLEX 
- FONT_HERSHEY_TRIPLEX, FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX

````python
image = np.zeros((1000,1000,3), np.uint8)
ourString =  'Hello World!'
cv2.putText(image, ourString, (155,290), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 3, (40,200,0), 4)
imshow("Messing with some text", image,4)
````
