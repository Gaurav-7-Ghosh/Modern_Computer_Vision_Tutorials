```python
import cv2
from matplotlib import pyplot as plt
```

## Creating a image display function with variable size 

```python
def imshow(title = "Image", image = None, size = 10):
    ## Getting the height and weight of the image in order to 
    ## maintain the aspect ratio of the display
    w, h = image.shape[0], image.shape[1] 
    aspect_ratio = w/h
    
    ## Calculating the image size by multiplying its current size with the aspect ratio
    ## in order to get corrected height according to current width
    plt.figure(figsize=(size * aspect_ratio,size))
    
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
nature_image = cv2.imread('./images/nature1.jpg')
```

```python
imshow("Reduced Aspect Ratio",nature_image,3)
```

```python
imshow("Increased Aspect Ratio",nature_image,13)
```

### GrayScaling Images

```python
gray_image = cv2.cvtColor(nature_image,cv2.COLOR_BGR2GRAY)
imshow("Gray_Image",gray_image)
```
