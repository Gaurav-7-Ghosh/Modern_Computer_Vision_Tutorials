## Loading Images

```python
nature_image = cv2.imread('./images/nature1.jpg')
```

## Syntax
#### cv2.imread(filename,flag)

##### <span style='color: red;'>filename:</span> The path to the image file.
##### <span style='color: red;'>flag:</span> The flag specifies the way how the image should be read.

##### cv2.<span style='color: blue;'>IMREAD_COLOR</span> – It specifies to load a color image. Any transparency of image will be neglected. It is the default flag. Alternatively, we can pass integer value 1 for this flag.

##### cv2.<span style='color: blue;'>IMREAD_GRAYSCALE</span> – It specifies to load an image in grayscale mode. Alternatively, we can pass integer value 0 for this flag. 

##### cv2.<span style='color: blue;'>IMREAD_UNCHANGED</span> – It specifies to load an image as such including alpha channel. Alternatively, we can pass integer value -1 for this flag.

### <span style='color: green;'>Return type:</span>  <span style='color: orange;'> Numpy Array </span> 


```python
print(nature_image) ## Returns a numpy array
```

## Displaying Images

```python
from matplotlib import pyplot as plt ## importing pyplot lib from matplotlib in order to display images
```

#### for displaying the image we have to understand that openCV by default reads images in the <span style= 'color:blue;'> BGR (Blue Green Red) color space</span> and not the more commonly used <span style= 'color:blue;'>RGB color space</span>

#### Hence we first use the function <span style= 'color:red;'> cvtColor</span> which is used to convert an image from one color space to another. It returns an image in the desired color space.

#### Then we use the <span style= 'color:red;'>imshow()</span> function in pyplot module of matplotlib library is used to display data as an image; i.e. on a 2D regular raster.
    


```python
plt.imshow(nature_image) ## Color Space not corrected so Blue--Red mapped
```

```python
plt.imshow(cv2.cvtColor(nature_image,cv2.COLOR_BGR2RGB)) ## Color space Corrected
```

## Creating a default function for displaying images read using cv2.imread

```python
def showimage(title="",image=None):
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
showimage("Natural Image",nature_image)
```

## Saving Images

#### <span style= 'color:red;'>cv2.imwrite() </span> method is used to save an image to any storage device. This will save the image according to the specified format in current working directory.

### Syntax

####  cv2.imwrite(filename, image)

### Parameters:
#### <span style= 'color:red;'> filename</span>: A string representing the file name. The filename must include image format like .jpg, .png, etc.
#### <span style= 'color:red;'> image:</span> It is the image that is to be saved. if only image_name.format is mentioned it is saved in the current working directory

#### <span style= 'color:red;'> Return Value: </span>It returns true if image is saved successfully.

```python
cv2.imwrite('output.png',nature_image)
```
